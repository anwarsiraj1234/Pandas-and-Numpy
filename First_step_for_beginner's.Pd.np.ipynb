# How many Data Structure in pandas?
Pandas has two types of data structures. 


```python
# a) Series-its a one dimensional array with indexex, it stores a single column or row of data in a Datafram.
# A one-dimensional array(labelled) capable of holding any type of data-Series. 
# Simple to say each row and column contain Uniform Data for example flort, or integers 
```

```python
# b) Data frame- It's tebular spreadsheet like structure representing rows each of which containsone or multiple columns. 
# A two-dimensional data (labeled) structure with columns of protentially different types of data - Dataframe.
# Simple to say each row and column contain different types of data for example , object 
```
